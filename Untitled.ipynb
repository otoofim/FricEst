```python
import torchvision.models as models
from torchvision import transforms
from frictionDataloader import *
import numpy as np
import torch
from torch.utils.data import DataLoader
from matplotlib import pyplot as plt
import torch.optim as optim
from tqdm import tqdm
import torchvision
import wandb
import os
import warnings
from statistics import mean
from UNet import *
```

```python
dataset_arg = {
        "season": "Winter",
        "date": "23-03-2021",
        "hour": "15:58",
        "year": "2021"
}
  

preprocess_in = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean = [0.485, 0.456, 0.406], std = [0.229, 0.224, 0.225]),
    transforms.Resize((256,256))
])


dataset = FrictionLoader(datasetConfig = dataset_arg, transform_in = preprocess_in)

train_loader = DataLoader(dataset = dataset, batch_size = 50, shuffle = False, drop_last = True)
```

```python
model = UNet(256)
model.load_state_dict(torch.load("./checkpoints/FriEst_friction1/best.pth")['model_state_dict'])
model = model.to(torch.device("cuda:0"))
```

```python
res = []
label = []
with tqdm(train_loader, unit="epoch") as epobar:
    for batch in epobar:
        out = model.inference(batch["image"].cuda())
        res.append(out.flatten())
        label.append(batch["label"])
```

```python
res_plot = torch.stack(res).flatten().detach().cpu().numpy()
```

```python
label_plot = torch.stack(label).flatten().detach().cpu().numpy()
```

```python
plt.figure(figsize=(12, 10), dpi=80)
plt.plot(res_plot, label = "estimated friction")
plt.plot(label_plot, label = "GT")
plt.legend()
plt.show()
```
